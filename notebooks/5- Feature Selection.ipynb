+ 2 feature selection methods were implemented in the paper; removing features with low variance and removing colinear variables.Both will be used here.

```python
import pandas as pd 
```

```python
df = pd.read_csv("ache_pchem.csv")
```

```python
df.head()
```

```python
X= df.drop(["Label"],axis=1)
```

```python
from sklearn.feature_selection import VarianceThreshold
```

```python
var = VarianceThreshold(0)
```

```python
var.fit(X)
```

```python
var.get_support()
```

```python
X = pd.DataFrame(var.transform(X),columns=X.columns[var.get_support()])
```

```python
X.head()
```

```python
df = pd.concat([X,df["Label"]],axis=1)
```

```python
df.head()
```

```python
df_sub = pd.read_csv("ache_sub.csv")
```

```python
X_ = df_sub.drop(["Label"],axis=1)
```

```python
X_.head()
```

```python
v = VarianceThreshold(0)
```

```python
X_ = pd.DataFrame(v.fit_transform(X_),columns=X_.columns[v.get_support()])
```

```python
X_.head()
```

```python
df_sub = pd.concat([X_,df_sub["Label"]],axis=1)
```

```python
df_sub
```

+ **Removing Multicolinearity**

Collinearity is a very common problem in machine learning projects. It is the correlation between the features of a dataset and it can reduce the performance of our models because it increases variance and the number of dimensions. It becomes worst when you have to work with unsupervised models.

```python
#!pip install collinearity
```

```python
from collinearity import SelectNonCollinear
```

```python
df.head()
```

```python
selector = SelectNonCollinear(0.7)
```

```python
X.head()
```

```python
y = df["Label"]
```

```python
selector.fit(X.values,y)
```

```python
mask = selector.get_support()
```

```python
df_pc = pd.DataFrame(selector.transform(X.values),columns=X.columns[selector.get_support()])
```

```python
df_pc.head()
```

```python
df_pc["Label"] = df["Label"]
```

```python
df_pc
```

```python
df_pc.to_csv("pubchem_dataset.csv",sep="\t",index=None)
```
