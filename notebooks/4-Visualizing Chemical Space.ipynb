```python
import pandas as pd 
import chemplot
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("dataset_std.csv",sep="\t")
```

```python
from chemplot import Plotter
```

```python
cp = Plotter.from_smiles(df["smiles"],target=df["label"],target_type="C")
```

```python
cp.pca()
```

```python
cp.visualize_plot()
```

```python
cp.tsne()
cp.visualize_plot()
```

```python
cp.umap()
cp.visualize_plot()
```

```python
cp = Plotter.from_smiles(df["smiles"],target=df["label"],target_type="C",sim_type="structural")
```
